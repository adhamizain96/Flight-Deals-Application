```python
from datetime import datetime, timedelta   
from data_manager import Data_Manager
from flight_search import Flight_Search
from notification_manager import Notification_Manager

data_manager = Data_Manager()
sheety_data = data_manager.get_destination_data()
flight_search = Flight_Search()
notification_manager = Notification_Manager()

starting_iata = 'ORD'
if sheety_data[0]['iataCode'] == '':
    for row in sheety_data:
        row['iataCode'] = flight_search.get_destination_code(row['city'])
    data_manager.destination_data = sheety_data

for travel_destination in sheety_data:
    flight = flight_search.check_flights(
        starting_iata,
        travel_destination['iataCode'],
        from_time = datetime.now(),
        to_time = datetime.now() + timedelta(days = 6*40)
    )
    if flight.price < travel_destination['lowestPrice']:
        notification_manager.send_msg(
            message = f"Check out this ✈️⬇️\nIt costs ${flight.price} to fly from {flight.origin_city} to {flight.destination_city}, from {flight.out_date} to {flight.return_date}."
        )
```
